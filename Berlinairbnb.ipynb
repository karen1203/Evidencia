```python
from funpymodeling.exploratory import freq_tbl
```

```python
berlin=pd.read_csv('berlin.csv')
berlin
```

```python
berlin1=berlin[['host_name','host_since','host_location','host_response_time','host_response_rate','host_listings_count','host_verifications','host_identity_verified','neighbourhood','property_type','room_type','accommodates','price','minimum_nights','maximum_nights','number_of_reviews','review_scores_accuracy','host_is_superhost','host_acceptance_rate','review_scores_cleanliness','review_scores_value','availability_365']]
berlin1
```

```python
berlin1.isnull().sum()
```

```python
t1=freq_tbl(berlin1['host_location'])
t1
```

```python
berlin1['host_location']=berlin1['host_location'].fillna('Berlin, Germany')
```

```python
t2=freq_tbl(berlin1['host_response_time'])
t2
```

```python
berlin1['host_response_time']=berlin1['host_response_time'].fillna('within an hour')
```

```python
t3=freq_tbl(berlin1['host_response_rate'])
t3
```

```python
berlin1['host_response_rate']=berlin1['host_response_rate'].fillna('100%')
```

```python
t4=freq_tbl(berlin1['neighbourhood'])
t4
```

```python
berlin1['neighbourhood']=berlin1['neighbourhood'].fillna('Berlin, Germany')
```

```python
t5=freq_tbl(berlin1['host_is_superhost'])
t5
```

```python
berlin1['host_is_superhost']=berlin1['host_is_superhost'].fillna('f')
```

```python
t6=freq_tbl(berlin1['host_acceptance_rate'])
t6
```

```python
berlin1['host_acceptance_rate']=berlin1['host_acceptance_rate'].fillna('100%')
```

```python
berlin1["review_scores_cleanliness"]=berlin1["review_scores_cleanliness"].fillna(round(berlin1["review_scores_cleanliness"].mean(),1))
```

```python
berlin1["review_scores_value"]=berlin1["review_scores_value"].fillna(round(berlin1["review_scores_value"].mode(),1))
```

```python
berlin1["review_scores_accuracy"]=berlin1["review_scores_accuracy"].fillna(round(berlin1["review_scores_accuracy"].mean(),1))
```

```python
berlin1.isnull().sum()
```

```python
berlin1['host_name']=berlin1['host_name'].fillna('Desconocido')
```

```python
t7=freq_tbl(berlin1['host_since'])
t7
```

```python
berlin1['host_since']=berlin1['host_since'].fillna('2021-12-03')
```

```python
berlin1["host_listings_count"]=berlin1["host_listings_count"].fillna(round(berlin1["host_listings_count"].mean(),0))
```

```python
t8=freq_tbl(berlin1['host_identity_verified'])
t8
```

```python
berlin1['host_identity_verified']=berlin1['host_identity_verified'].fillna('t')
```

```python
berlin1["review_scores_value"]=berlin1["review_scores_value"].fillna(round(berlin1["review_scores_value"].mean(),2))
```

```python
berlin1.isnull().sum()
```

```python
cuali=berlin1[['host_name','host_acceptance_rate','host_since','host_location','host_response_time','host_response_rate','host_listings_count','host_verifications','host_identity_verified','neighbourhood','property_type','room_type','accommodates','price','minimum_nights','maximum_nights','number_of_reviews','host_is_superhost']]
```

```python
cuanti=berlin1[['host_listings_count','accommodates','minimum_nights','maximum_nights','number_of_reviews','review_scores_accuracy','review_scores_cleanliness','review_scores_value','availability_365']]
```

```python
plt.boxplot(cuanti)
```

```python
q1 = cuanti['host_listings_count'].quantile(0.25)
q2 = cuanti['host_listings_count'].quantile(0.75)
iqr = q2 - q1

lower_bound1 = q1 - 1.5 * iqr
upper_bound1 = q2 + 1.5 * iqr
cuanti = cuanti[(cuanti['host_listings_count'] >= lower_bound1) & (cuanti['host_listings_count'] <= upper_bound1)]
```

```python
cuanti['host_listings_count'] = np.where(cuanti['host_listings_count'] > upper_bound1, upper_bound1, cuanti['host_listings_count'])

cuanti['host_listings_count'] = np.where(cuanti['host_listings_count'] < lower_bound1, lower_bound1, cuanti['host_listings_count'])
```

```python
q11 = cuanti['minimum_nights'].quantile(0.25)
q21 = cuanti['minimum_nights'].quantile(0.75)
iqr1 = q21 - q11

lower_bound11 = q11 - 1.5 * iqr1
upper_bound11 = q21 + 1.5 * iqr1
cuanti = cuanti[(cuanti['minimum_nights'] >= lower_bound11) & (cuanti['minimum_nights'] <= upper_bound11)]
```

```python
cuanti['minimum_nights'] = np.where(cuanti['minimum_nights'] > upper_bound11, upper_bound11, cuanti['minimum_nights'])

cuanti['minimum_nights'] = np.where(cuanti['minimum_nights'] < lower_bound11, lower_bound11, cuanti['minimum_nights'])
```

```python
q12 = cuanti['maximum_nights'].quantile(0.25)
q22 = cuanti['maximum_nights'].quantile(0.75)
iqr2 = q22 - q12

lower_bound12 = q12 - 1.5 * iqr2
upper_bound12 = q22 + 1.5 * iqr2
cuanti = cuanti[(cuanti['maximum_nights'] >= lower_bound12) & (cuanti['maximum_nights'] <= upper_bound12)]
```

```python
cuanti['maximum_nights'] = np.where(cuanti['maximum_nights'] > upper_bound12, upper_bound12, cuanti['maximum_nights'])

cuanti['maximum_nights'] = np.where(cuanti['maximum_nights'] < lower_bound12, lower_bound12, cuanti['maximum_nights'])
```

```python
q13 = cuanti['number_of_reviews'].quantile(0.25)
q23 = cuanti['number_of_reviews'].quantile(0.75)
iqr3 = q23 - q13

lower_bound13 = q13 - 1.5 * iqr3
upper_bound13 = q22 + 1.5 * iqr3
cuanti = cuanti[(cuanti['number_of_reviews'] >= lower_bound13) & (cuanti['number_of_reviews'] <= upper_bound13)]
```

```python
cuanti['number_of_reviews'] = np.where(cuanti['number_of_reviews'] > upper_bound13, upper_bound13, cuanti['number_of_reviews'])

cuanti['number_of_reviews'] = np.where(cuanti['number_of_reviews'] < lower_bound13, lower_bound13, cuanti['number_of_reviews'])
```

```python
berlimp=pd.concat([cuanti,cuali], axis=1)
berlimp
```
